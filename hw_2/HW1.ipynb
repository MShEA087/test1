```python
class Chocolate:
    def __init__(self, price: float, topping: str = None):
        if price <= 0:
            raise ValueError("Цена должна быть больше нуля")
        self.price = price
        self.topping = topping

    def show_my_chocolate(self):
        if self.topping:
            print(f"Шоколадный батончик, добавка: {self.topping}")
        else:
            print("Обычный шоколадный батончик")

    def __str__(self):
        return f"Шоколадный батончик, {self.price} рублей"
```

```python
choco1 = Chocolate (150.00)
print (choco1)
choco1.show_my_chocolate()
```

```python
choco2 = Chocolate (50.00, "орехи")
choco2.show_my_chocolate()
print (choco2)
```

```python
choco3 = Chocolate (75.00, "корица")
choco3.show_my_chocolate()
print (choco3)
```

```python
choco4 = Chocolate (-15.00)
print (choco4)
```
