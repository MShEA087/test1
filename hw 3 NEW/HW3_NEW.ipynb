Исправлено

```python
df['Gender_Name'] = df['Name'] + ' (' + df['Sex'] + ')'

# Группировка по классу и полу и нахождение самой популярной имени
most_popular_names = df.groupby(['Pclass', 'Sex'])['Name'].agg(lambda x: x.mode()[0])

# Вывод результатов
for pclass in most_popular_names.index.levels[0]:
    male_name = most_popular_names[pclass]['male'] if 'male' in most_popular_names[pclass] else None
    female_name = most_popular_names[pclass]['female'] if 'female' in most_popular_names[pclass] else None

    print(f'Класс {pclass}:')
    print(f'  Самое популярное мужское имя: {male_name}')
    print(f'  Самое популярное женское имя: {female_name}')
```
