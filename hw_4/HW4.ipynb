Постройте график типа pie chart для переменных Survived, Pclass, подпишите доли в процентах

```python
plt.figure(figsize=(12, 6))

# Подсчет долей выживших и невыживших
survived_counts = df['Survived'].value_counts()
plt.subplot(1, 2, 1)  # 1 строка, 2 столбца, 1-я диаграмма
plt.pie(survived_counts, labels=['Not Survived', 'Survived'], autopct='%1.1f%%', startangle=90)
plt.title('Survival')

# Построение круговой диаграммы для переменной Pclass
pclass_counts = df['Pclass'].value_counts()
plt.subplot(1, 2, 2)  # 1 строка, 2 столбца, 2-я диаграмма
plt.pie(pclass_counts, labels=pclass_counts.index, autopct='%1.1f%%', startangle=80)
plt.title('Passenger Class')

# Отображение графиков
plt.tight_layout()
plt.show()
```

Постройте интерактивный sunburst plot (визуализация иерархических данных) с помощью plotly. На первом уровне иерархии - количество пассажиров в каждом из классов, а на втором количество женщин/мужчин в этом классе.

```python
import plotly.express as px

# Подсчет количества пассажиров по классам и полу
data = df.groupby(['Pclass', 'Sex']).size().reset_index(name='Count')

# Создание sunburst plot
fig = px.sunburst(data, path=['Pclass', 'Sex'], values='Count',title='Количество пассажиров по классам и полу')
plt.legend(title='Пол')
fig.show()
```
